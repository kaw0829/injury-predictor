###  Utilities


```python
def catboost_enc(X_train,y_train,X_test, features, rs):
    # encode nonnumerical cols listed in features using training data
    cb_enc = ce.CatBoostEncoder(cols=features)
    cb_enc.fit(X_train[features], y_train)
    # add columns for encoded data and join them to the correct col
    X_test = X_test.join(cb_enc.transform(X_test[features]).add_suffix('_cb'))
    # drop categorical cols
    X_test.drop(features, axis=1, inplace=True)
    return  X_test
```

```python
def extractDayOfTeamYear(dateObj):
  dateofyear = float(dateObj.strftime("%j"))
  dayOfTeamYear = ((dateofyear - 62) % 365)
  return dayOfTeamYear
```

```python
def createInjuryDF(team, player, injury, dateInjured):
    teamRoster = roster[roster.Team == teams_dict[f'{team}']]
    rosterEntry = teamRoster[teamRoster.Player ==player]
    playerInfo = rosterEntry.iloc[0]
    todaysDate = datetime.datetime.today()
    dateInjDT = datetime.datetime(dateInjured.year, dateInjured.month, dateInjured.day)
    birthDate1 = dateInjDT-rosterEntry.iloc[0].BirthDate
    birthdate2 = float(birthDate1.days)
    df = pd.DataFrame(columns=['Team','Year','Injury_Category','Age' ,'Pos','Wt','Ht', 'Month_Injured','Day_Of_Team_Year'])
    df.loc[0] = [playerInfo['Team'], 2021.0, injury,birthdate2, rosterEntry.iloc[0].Pos,rosterEntry.iloc[0].Wt,rosterEntry.iloc[0].Ht,  int(dateInjured.month), extractDayOfTeamYear(dateInjured), ]
    return df

```

```python
def predictInjuryDuration(playerDF):
  X_test = catboost_enc(X_train, y_train, playerDF, features, RANDOM_SEED)
  preds =  cbmodel.predict(X_test)
  return preds[0]
```

```python
def mostSimilarInjury(playerDF):
  df = Inj_Bio[Inj_Bio['Injury_Category'] == playerDF['Injury_Category'][0]]
  if df.shape[0] > 1:
    df = similarityScore(df, playerDF)
    df.iloc[0]
    return df.iloc[0:10]
  else: 
    return df
  
  
```

```python
def similarityScore(dataframe, playerDF):
  def playerDifference(colVal, colLabel):
    return abs(colVal - playerDF[colLabel])
   
  
  dataframe['ageDif_'] = dataframe['Age'].apply(lambda x: playerDifference(x, 'Age'))
  dataframe['wtDif_'] = dataframe['Wt'].apply(lambda x: playerDifference(x, 'Wt'))
  dataframe['HtDif_'] = dataframe['Ht'].apply(lambda x: playerDifference(x, 'Ht'))
  ageStdDiv = dataframe.Age.std()
  wtStdDiv = dataframe.Wt.std()
  htStdDiv = dataframe.Ht.std()
  ageImportance = .0475
  wtImportance  = .0350
  htImportance  = .0206
  similarityScoreCol = ((dataframe['ageDif_'] / ageStdDiv) / ageImportance + (dataframe['wtDif_'] / wtStdDiv) / wtImportance + (dataframe['HtDif_'] / htStdDiv / htImportance))

  dataframe['similarityScore'] = similarityScoreCol
  dataframe = dataframe.sort_values(by=['similarityScore'])
  return dataframe
```

```python
def statAnalysisReport(y_train):
  yStdDev = y_train.Days_Missed.std()
  yMean = y_train.Days_Missed.mean()
  return yStdDev, yMean
  
```

### 3. Reports
Select a team, player, injury type and date injured to generate a report with prediction of injury time.

```python
initTeam = roster[roster.Team == teams_dict[f'Arizona Cardinals']] 
initPlayers = initTeam["Player"]  

teamSelector = widgets.Dropdown(
                options=teams,
                value='Arizona Cardinals',
                description='Team:',
                disabled=False,
                );

playerSelector = widgets.Dropdown(
                options = initPlayers,
                value=initPlayers[0],
                decription="Player:",
                disabled =True
                );

def select_team(team):
  team_ = roster[roster.Team == teams_dict[f'{team}']]
  playerSelector.disabled = False
  playerSelector.options = team_["Player"]
   
#def select_player(player):

i = widgets.interactive(select_team, team=teamSelector)

injurySelector = widgets.Dropdown(
      options = injuryCats,
      decscription = "Injurys:",
      disabled=False
      );
dp = widgets.DatePicker(
      description='Pick a Date',
      disabled=False,
      value = datetime.date(2021,7,1)
      )

output = widgets.Output(layout={'border': '1px solid black', 'width': '200%'})
button = widgets.Button(description="Submit")

def on_button_clicked(b):
  output.clear_output()

  playerDF   = createInjuryDF(teamSelector.value,playerSelector.value,injurySelector.value,dp.value)
  # mSI        = mostSimilarInjury(playerDF)
  
  injuryPred = predictInjuryDuration(playerDF)
  mSI        = mostSimilarInjury(playerDF)
  sI         =  mSI.iloc[0]
  yStdDev, yMean = statAnalysisReport(y_train)
  
  mSIMean = mSI.Days_Missed.mean()
  illnessdf = Inj_Bio[Inj_Bio['Injury_Category'] == injurySelector.value]
  averageDaysMissed = illnessdf.Days_Missed.mean()
  stdADM = illnessdf.Days_Missed.std()
  MAE = 13
  daysMissedData = {
                    'Predicted': injuryPred,
                    'Most Similar Player': sI['Days_Missed'],
                    'Mean 10 most Similar': mSIMean,
                    f'mean {injurySelector.value}': averageDaysMissed,
                    'Mean all injuries': yMean, 
                    }
 
  with output: 
      newLine = '\n'
      print(f'Standard deviation for all injuries is {yStdDev:.2f} and the mean days missed is {yMean:.2f}' )
      print(f'{playerSelector.value} sustained a/an {injurySelector.value} injury and is estimated to miss {int(injuryPred)} days')
      print(f'The most similar injury recorded was {sI["Name"]}, AGE:{int(sI["Age"] / 365)}, WT:{sI["Wt"]}lbs,  HT:{sI["Ht"]} inches, Games Missed: {int(sI["Days_Missed"])}{newLine}')
      print(f'The standard deviation of days missed for {injurySelector.value} is {stdADM:.2f} and Mean is {averageDaysMissed:.2f}')
      print(f'We can expect recovery time to fall between {int(injuryPred - MAE)} days and {int(injuryPred + MAE)} days')

     
      plt.figure(figsize=(10,6))
      plt.bar(range(len(daysMissedData)), list(daysMissedData.values()), align='center')
      plt.xticks(range(len(daysMissedData)), list(daysMissedData.keys()))
      plt.show();



button.on_click(on_button_clicked)

left_box = widgets.VBox([i, playerSelector, button])
right_box = widgets.VBox([injurySelector,dp, output])
widgets.HBox([left_box,right_box])
```

### 4. Graphs


Select an injury category and press the submit button to graph that injury category against 
the age of the players in days. <br> Age is one of the more statisically important
features, the red line is a regression line.  <br>For many injuries as age increases
so does the recovery time as would be expected.

```python
injurySelector = widgets.Dropdown(
      options = injuryCats,
      decscription = "Injurys:",
      disabled=False
      );

button = widgets.Button(description="Button")

out = widgets.Output()




def on_button_clicked(b):
    with out:
        clear_output(True)
        # plot([1,2],[2,1])
        fig, ax = plt.subplots()
        df = Inj_Bio[Inj_Bio['Injury_Category'] == injurySelector.value]
        m, b = np.polyfit(df['Age'],df['Days_Missed'] , 1)
        ax.set_title(f'Regression line for Age/{injurySelector.value} Injury')
        ax.set_xlabel('Age in Days')
        ax.set_ylabel('Days Missed')
        ax.plot(df['Age'], m*df['Age'] + b, color='red'  )
        ax.scatter(df['Age'], df['Days_Missed']);
        with plt.style.context('seaborn'):
          plt.style.use('seaborn')
          plt.show()
button.on_click(on_button_clicked)

display(injurySelector)
display(button)
out

```

```python
def plot_feature_Importance(columns, importances, n=10):
  """[creates a horizontal bar graph visualizing the importance of each feature col]

  Args:
      columns ([list]): [columns in df]
      importances ([array]): [array with numerical importance of each col, 1 being most significant, 0 being insignificant]
      n (int, optional): [no of features to chart]. Defaults to 10.
  """
  df = (pd.DataFrame({"features": columns,
                      "feature_importances": importances})
        .sort_values("feature_importances", ascending=False)
        .reset_index(drop=True))
  
  #plot the df
  fig, ax = plt.subplots()
  ax.barh(df["features"][:n], df["feature_importances"][:n])
  ax.set_ylabel("Features")
  ax.set_xlabel("Feature importance")
  ax.invert_yaxis()
```

This graph displays the importance of each feature as determined by the CatBoostModel.  <br><b>(Data is gathered from PreProcessing_Injury_Predictor)

```python
plot_feature_Importance(feature_List, feature_importance)
```
