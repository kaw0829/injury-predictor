```python
def parse_ht(ht):
    # format: 7' 0.0"
    ht_ = ht.split("-")
    ft_ = float(ht_[0])
    in_ = float(ht_[1])
    return (12*ft_) + in_
```

```python
def fixPos(P):
  if P == 'LT' or P == 'RT' or P == 'OT':
    P = 'T'
  elif P == 'EDGE':
    P = 'DE'
  elif P == 'G' or P == 'C' or P == 'LG' or P == 'RG' or P == 'G/C':
    P = 'GC'
  elif "LB" in P:
    P = "LB"
  elif 'RB' in P or P == 'HB':
    P = 'RB'
  elif P == 'NT':
    P = 'DT'
  elif 'TE' in P:
    P = 'TE'
  elif 'WR' in P:
    P = 'WR'
  elif P == 'FS' or P == "SS":
    P = 'S'
  elif P == 'CB' or P == 'PR':
    P = 'DB'
  elif P == 'K' or P == 'P':
    P = 'PK'
  return P
```

```python
def parse_name(Player):
    if '\\' in Player:
        plyr_ = Player.split("\\")
        name_ = str(plyr_[0])
    else:
      name_ = str(Player)
    reg1 = r'(?<=^.).*\s'
    newname = re.sub(reg1, ' ', name_)
    return newname

```

```python
def formatRosterData(data):
  data.drop('Drafted', axis=1, inplace=True)
  data.drop('College', axis=1, inplace=True)
  data['No'].fillna("0", inplace=True)
  data['GS'].fillna("0", inplace=True)
  data.dropna(axis=0, inplace=True)
  data.loc[data['Yrs'] == 'Rook', 'Yrs'] = '0'
  data["Ht"] = data["Ht"].apply(lambda x:parse_ht(x))
  data["Player"] = data['Player'].apply(lambda x: parse_name(x))
  data["Pos"] = data['Pos'].apply(fixPos)
  return data
```

```python


def MergeIR(data):
  df = pd.DataFrame()
  injuryCategory = 0
  dateInjured = np.datetime64

  for index, row in data.iterrows():
    if(row[6] == 1):
     
      mask = ((df["Date_Injured"] < row[2]) & (df['Year'] == np.float64(row[4])))
      dataWithMask = df.loc[mask].sort_values(by='Date_Injured', ascending=False)

      
      Flag = True
      for index2, row2 in dataWithMask.iterrows():
        if row[0] == row2[0] and row[3] == row2[3] and Flag == True:
          dateInjured = row2[1]
          injuryCategory = row2[5]
          d = {'Name': row[0], 'Date_Injured':dateInjured, 'Date_Recovered':row[2], "Team":row[3], "Year":row[4], 'Injury_Category':injuryCategory, 'Keep':row[6]}
          ser = pd.Series(data=d, index=['Name', 'Date_Injured', 'Date_Recovered', 'Team', 'Year', 'Injury_Category', 'Keep'])
          df = df.append(ser, ignore_index=True)
          Flag = False
    else:
      df = df.append(row, ignore_index=True)
  return df
```

```python
def formatName1(Player):
    if ' ' in Player:
        plyr_ = Player.split(" ")
        fname_ = str(plyr_[0])
        lname_ = str(plyr_[1])
        name_ = f'{fname_} {lname_}'
    else:
      name_ = str(Player)
    reg1 = r'(?<=^.).*\s'
    newname = re.sub(reg1, ' ', name_)
    return newname
```

```python
def mergeData(injurydf, rosterdf):
  injurydf['Name'] = injurydf['Name'].apply(lambda x: formatName1(x))
  injurydf['Year'] = pd.to_numeric(injurydf["Year"], downcast='integer')
  rosterdf['Year'] = pd.to_numeric(rosterdf["Year"], downcast='integer')
  rosterdf['Name'] = rosterdf['Player']
  rosterdf.drop('Player', axis = 1, inplace=True)
  rosterdf['mergeKey'] = rosterdf['Name'] + rosterdf['Team'] + rosterdf['Year'].astype(str)
  injurydf['mergeKey'] = injurydf['Name'] + injurydf['Team'] + injurydf['Year'].astype(str)
  
  primaryDF = pd.merge(injurydf, rosterdf, on='mergeKey', how='left')
  return primaryDF
```

```python
df = mergeData(injurydf, rosterdf)
```

## Cleanup post merge

```python
def cleanUpMergeData(df):
  #drop redundant columns
  #df.Date_Injured = df.Age.astype(np.int64)
  df = df.dropna(axis=0)
  df = df.drop(['Team_y', 'Year_y', 'Name_y', 'Keep', 'mergeKey','No'], axis=1)
  #Rename columns after merge
  df = df.rename(columns={'Name_x': 'Name', 'Team_x':'Team', 'Year_x':'Year'})
  # calculates age at time of injury using birthdate
  df['Age'] = ((pd.to_datetime(df['Date_Injured']) - pd.to_datetime(df['BirthDate'])).dt.days).astype(np.float64)
  # adds a days missed cat which wil be the target category
  df['Days_Missed'] = ((pd.to_datetime(df['Date_Recovered']) - pd.to_datetime(df['Date_Injured'])).dt.days).astype(np.float64)
  #drops columns that are not useful for model and are missing data
  df = df.drop(['G', 'GS', 'Yrs', 'AV'], axis=1)
  # drops data before 2009 as IR rules in the nfl did not allow players to return
  df = df[df['Year'] > 2008]
  return df
```
